All the data is standard and in consistent format except for file 1 - January. 
Standardize file 1 to match all other files and drop columns that do not match all other files.

```python
#veiw January file again. 

Jan.head()
```

```python
#drop tripduration, birth year, bikeid, and gender

Jan_drop = Jan.drop(columns=['tripduration', 'bikeid', 'birth year', 'gender'])
Jan_drop.head()
```

```python
#insert missing columns in January file so the amount of columns match. 
Jan_drop.insert(1,'ride_id', 'None')



```

```python
Jan_drop.insert(1, 'rideable_type', 'None')


Jan_drop.head()
```

```python
#rename the columns to match other file columns 

Jan_rename = Jan_drop.rename(columns={"starttime":"started_at","stoptime":"ended_at",
                                     "start station id":"start_station_id", "start station name":"start_station_name",
                                     "start station latitude":"start_lat","start station longitude":"start_lng",
                                      "end station id":"end_station_id","end station name":"end_station_name",
                                     "end station latitude":"end_lat","end station longitude":"end_lng",
                                      "usertype":"member_casual"})
Jan_rename.head()
```

```python
#change column order to match other files

Jan_neworder = Jan_rename[['ride_id', 'rideable_type','started_at','ended_at','start_station_name','start_station_id',
                          'end_station_name','end_station_id','start_lat','start_lng','end_lat','end_lng','member_casual']]

Jan_neworder.head()
```

```python
#veiw the new order against one of the other dataframes

Dec.head()
```

```python
#replace "customer to casual", and "subscriber to member" in January file/df

Jan_neworder['member_casual'] = Jan_neworder['member_casual'].replace(['Customer','Subscriber'],['casual','member',])

Jan_neworder.head()
```

```python
#append all the dataframes together and view

# Jan_neworder
# Feb 
# Mar 
# Apr
# May 
# Jun 
# Jul 
# Aug 
# Sept 
# Oct 
# Nov 
# Dec 

files_append = Jan_neworder.append(Feb).append(Mar).append(Apr).append(May).append(Jun).append(Jul).append(Aug).append(Sept).append(Oct).append(Nov).append(Dec)

files_append.head()
```

```python
#view data

files_append.tail()
```
